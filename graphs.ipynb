```python
class Graph:
    def __init__(self):
        self.graph = {}

    def add_edge(self, u, v):
        if u not in self.graph:
            self.graph[u] = []
        if v not in self.graph:
            self.graph[v] = []
        self.graph[u].append(v)
        self.graph[v].append(u)

    def bfs(self, start):
        visited = {}
        queue = []

        for vertex in self.graph:
            visited[vertex] = False

        visited[start] = True
        queue.append(start)

        while queue:
            vertex = queue.pop(0)
            print(vertex, end=' ')

            for neighbor in self.graph[vertex]:
                if not visited[neighbor]:
                    visited[neighbor] = True
                    queue.append(neighbor)

    def dfs(self, start):
        visited = {}

        for vertex in self.graph:
            visited[vertex] = False

        def dfs_helper(vertex):
            visited[vertex] = True
            print(vertex, end=' ')

            for neighbor in self.graph[vertex]:
                if not visited[neighbor]:
                    dfs_helper(neighbor)

        dfs_helper(start)


# Example usage:
if __name__ == "__main__":
    g = Graph()
    g.add_edge(0, 1)
    g.add_edge(0, 2)
    g.add_edge(1, 2)
    g.add_edge(2, 0)
    g.add_edge(2, 3)
    g.add_edge(3, 3)

    print("BFS starting from vertex 2:")
    g.bfs(2)

    print("\nDFS starting from vertex 2:")
    g.dfs(2)
```

```python
class Graph:
    def __init__(self):
        self.graph = {}

    def add_edge(self, u, v, w):
        if u not in self.graph:
            self.graph[u] = []
        if v not in self.graph:
            self.graph[v] = []
        self.graph[u].append((v, w))
        self.graph[v].append((u, w))

    def dfs(self, start):
        visited = [start]
        stack = [start]
        result = []

        while stack:
            top = stack.pop()
            result.append(top)
            
            for neighbor, _ in self.graph[top]:
                if neighbor not in visited:
                    visited.append(neighbor)
                    stack.append(neighbor)

        return result

    def bfs(self, start):
        visited = [start]
        queue = [start]
        result = []

        while queue:
            top = queue.pop(0)
            result.append(top)

            for neighbor, _ in self.graph[top]:
                if neighbor not in visited:
                    visited.append(neighbor)
                    queue.append(neighbor)

        return result


# Example usage:
if __name__ == '__main__':
    g = Graph()
    g.add_edge(0, 1, 4)
    g.add_edge(0, 2, 8)
    g.add_edge(0, 3, 8)
    g.add_edge(4, 5, 11)
    g.add_edge(4, 6, 7)
    g.add_edge(4, 7, 2)
    g.add_edge(0, 4, 4)



    print("DFS starting from vertex 0:")
    dfs_result = g.dfs(0)
    print(dfs_result)

    print("\nBFS starting from vertex 0:")
    bfs_result = g.bfs(0)
    print(bfs_result)
```
