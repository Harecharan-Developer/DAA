# Brute Force

```python
def bruteForce(arr):
    maxSum = -float('inf')
    n=len(arr)
    for i in range(n):
        temp = 0
        for j in range(i,n):
            temp+=arr[j]

            if temp>maxSum:
                maxSum=temp
    
    print(f"The maximum sum is: {maxSum}")
```

```python
arr = [-2, -3, 4, -1, -2, 1, 5, -3]
bruteForce(arr)
```

```python
def maxCrossingSum(arr,low,mid,high):
    leftSum = -float('inf')
    temp = 0
    for i in range(mid,low-1,-1):
        temp+=arr[i]
        if temp>leftSum:
            leftSum=temp
    
    rightSum = -float('inf')
    temp = 0
    for i in range(mid+1,high+1):
        temp+=arr[i]
        if temp>rightSum:
            rightSum=temp
    
    return leftSum+rightSum




def maxSubArraySum(arr,low,high): #divide and conquer

    if low==high:
        return arr[low]
    else:
        mid = (low+high)//2
        left = maxSubArraySum(arr,low,mid)
        right = maxSubArraySum(arr,mid+1,high)
        crossing = maxCrossingSum(arr,low,mid,high)
        return max(left,right,crossing)

#driver code
arr = [-2, -3, 4, -1, -2, 1, 5, -3]
maxSum = maxSubArraySum(arr,0,len(arr)-1)
print(f"The maximum sum is: {maxSum}")
```

```python
def maxSubArraySumdp(arr):
    dp = [0]*len(arr)
    dp[0] = arr[0]
    res = arr[0]

    n= len(arr)

    for i in range(1,n):
        dp[i] = max(arr[i],arr[i]+dp[i-1])
        res = max(res,dp[i])
    
    print(f"The maximum sum is: {res}")
#driver code
arr = [-2, -3, 4, -1, -2, 1, 5, -3]
maxSubArraySum(arr)
```

# positive values

```python
import timeit
import random
import matplotlib.pyplot as plt

def generate_random_array(size):
    return [random.randint(0, 1000) for _ in range(size)]



input_sizes = [1000, 2000, 3000, 4000, 5000]
execution_times_bruteforce = []
execution_times_dac = []
execution_times_dp = []

for size in input_sizes:
    arr = generate_random_array(size)
    bftime = timeit.timeit(lambda: bruteForce(arr), number=1)
    dctime = timeit.timeit(lambda: maxSubArraySum(arr), number=1)
    dp = timeit.timeit(lambda: maxSubArraySumdp(arr), number=1)
    execution_times_bruteforce.append(bftime)
    execution_times_dac.append(dctime)
    execution_times_dp.append(dp)

plt.plot(input_sizes, execution_times_bruteforce, label="Brute Force")
plt.plot(input_sizes, execution_times_dac, label="Divide and conquer")
plt.plot(input_sizes, execution_times_dp, label="Dynamic Programming")
plt.xlabel("Input Size")
plt.ylabel("Execution Time")
plt.title("Time Complexity")
plt.legend()
plt.show()
```
