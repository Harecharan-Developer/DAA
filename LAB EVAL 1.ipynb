### Merge Sort

Merge Sort is a divide-and-conquer sorting algorithm that consistently divides the input data into smaller halves, sorts them, and then merges them back together. It is known for its stable performance and guaranteed O(n log n) time complexity, making it efficient for large datasets.

#### Performance Analysis
- **Stability**: Merge Sort is a stable sorting algorithm, meaning it maintains the relative order of equal elements.
- **Time Complexity**: It guarantees a time complexity of O(n log n) in the worst, average, and best cases.
- **Space Complexity**: Merge Sort requires additional memory for the merging step, resulting in a space complexity of O(n).

#### Use Cases
Merge Sort is suitable for scenarios where stability and predictable performance are crucial. It excels with large datasets and is commonly used for external sorting tasks.

### Insertion Sort
#### Overview
Insertion Sort is a simple comparison-based sorting algorithm that builds the final sorted array one item at a time. It is known for its simplicity and suitability for small datasets.

#### Performance Analysis
- **Stability**: Insertion Sort is a stable sorting algorithm.
- **Time Complexity**: In the worst case, it has a time complexity of O(n^2), making it less efficient for large datasets.
- **Space Complexity**: Insertion Sort is an in-place sorting algorithm, meaning it has a space complexity of O(1).

#### Use Cases
Insertion Sort is most effective for small datasets or nearly sorted data due to its simplicity and low overhead. It is often used as a building block in more complex algorithms.

### Custom Sort (new_sort)
#### Overview
Custom Sort (new_sort) is a sorting algorithm with a customizable parameter k. It divides the input into subarrays of size k, sorts them, and then merges them together. The value of k can significantly impact its performance.

#### Performance Analysis
- **Stability**: Custom Sort is a stable sorting algorithm.
- **Time Complexity**: The time complexity depends on the value of k. For k = 1, it has a time complexity of O(n^2), but for larger k values, it can be more efficient, approaching O(n log n).
- **Space Complexity**: It has a space complexity of O(n) due to the merging step.

#### Use Cases
Custom Sort (new_sort) is versatile and suitable for various scenarios. By adjusting the value of k, it can adapt to different input sizes and distribution patterns. For small k values, it is similar to Insertion Sort, while larger k values approach the efficiency of Merge Sort.

### Comparison and Conclusion
- Merge Sort offers stable performance with a guaranteed O(n log n) time complexity, making it ideal for large datasets.
- Insertion Sort is simple and efficient for small datasets or nearly sorted data but becomes inefficient for larger inputs.
- Custom Sort (new_sort) is adaptable based on the value of k, making it suitable for a wide range of scenarios. Smaller k values mimic Insertion Sort, while larger k values approach the efficiency of Merge Sort.

The choice of sorting algorithm depends on the specific requirements of the task. Merge Sort is a reliable choice for large datasets, Insertion Sort for small or nearly sorted data, and Custom Sort with an optimal k value for versatility and adaptability.

# MERGE SORT ALGORITHM

````python
def merge_sort(array):
    if len(array)>1:
        mid=len(array)//2
        #splitting array into left half and right half
        left_array=array[:mid]
        right_array=array[mid:]
        merge_sort(left_array)
        merge_sort(right_array)

        #merging step
        i,j,k=0,0,0

        while i<len(left_array) and j<len(right_array):
            if left_array[i]<right_array[j]:
                array[k]=left_array[i]
                i+=1
            else:
                array[k]=right_array[j]
                j+=1

            k+=1

        while i<len(left_array):
            array[k]=left_array[i]
            i+=1
            k+=1

        while j<len(right_array):
            array[k]=right_array[j]
            j+=1
            k+=1
````

# INSERTION SORT ALGORITHM

````python
def insertion_sort(array):
    n=len(array)
    for i in range(1,n):
        key=array[i]
        j=i-1
        while j>=0 and key<array[j]:
            array[j+1]=array[j]
            j-=1
        array[j+1]=key
````

# SPLIT SORT MERGE ALGORITHM

````python
def new_sort(array,k):
    n=len(array)
    subArraySize=k
    if n%k:
        subArraySize+=1
    subArrays=[]
    for i in range(0,n,subArraySize):
        subArrays.append(array[i:i+subArraySize])

    for subarray in subArrays:
        insertion_sort(subarray)

    sorted_arr=[]
    while len(subArrays)>0:
        minval=[]
        for subarray in subArrays:
            minval.append(subarray[0])

        minvalindex=minval.index(min(minval))
        sorted_arr.append(subArrays[minvalindex].pop(0))

        if not subArrays[minvalindex]:
            subArrays.pop(minvalindex)

    return sorted_arr
````

````python
import random
import matplotlib.pyplot as plt
import time

# Define your merge_sort, insertion_sort, and new_sort functions here

input_sizes = list(range(1, 10001, 100))
execution_times_merge = []
execution_times_insertion = []

k_values = [1, 2, int(len(input_sizes) ** 0.5), len(input_sizes) // 2]
execution_times_new = {k: [] for k in k_values}

for size in input_sizes:
    data = [random.randint(1, 1000) for _ in range(size)]

    start_time = time.time()
    merge_sort(data.copy())
    execution_time_merge = time.time() - start_time
    execution_times_merge.append(execution_time_merge)

    start_time = time.time()
    insertion_sort(data.copy())
    execution_time_insertion = time.time() - start_time
    execution_times_insertion.append(execution_time_insertion)

    for k in k_values:
        start_time = time.time()
        new_sort(data.copy(), k)
        execution_time_custom = time.time() - start_time
        execution_times_new[k].append(execution_time_custom)

# Plot the results
plt.plot(input_sizes, execution_times_merge, label='Merge Sort')
plt.plot(input_sizes, execution_times_insertion, label='Insertion Sort')

for k in k_values:
    plt.plot(input_sizes, execution_times_new[k], label=f'Custom Sort (k={k})')

plt.xlabel('Input Size')
plt.ylabel('Execution Time (s)')
plt.legend()
plt.title('Comparison of Sorting Algorithms')
plt.show()
````

<h2>Overview</h2>
<p>
        This documentation describes an algorithm to find the Minimum Spanning Tree (MST) of a graph while excluding non-bridge edges with the maximum weight. The process involves iterative removal of non-bridge edges and updating the list of bridges in the graph.
    </p>

 <h2>Algorithm Steps</h2>
    <ol>
        <li>
            <h3>Find All Bridges</h3>
            <p>
                Start by identifying all the bridge edges in the given graph using a bridge-finding algorithm. Bridges are edges whose removal would increase the number of connected components in the graph.
            </p>
        </li>
        <li>
            <h3>Iterative Process</h3>
            <p>
                Perform an iterative process to construct the Minimum Spanning Tree (MST) while excluding non-bridge edges.
                <br>
                In each iteration:
            </p>
            <ul>
                <li>
                    <p>
                        Check the edge with the maximum weight in the current graph.
                    </p>
                </li>
                <li>
                    <p>
                        Determine whether this edge is a bridge or not by comparing it to the list of previously identified bridges.
                    </p>
                </li>
                <li>
                    <p>
                        If the edge is not a bridge (i.e., its removal does not disconnect the graph), remove it from the graph.
                    </p>
                </li>
                <li>
                    <p>
                        If the edge is a bridge, keep it in the MST.
                    </p>
                </li>
                <li>
                    <p>
                        Continue this process until no non-bridge edges can be removed.
                    </p>
                </li>
            </ul>
        </li>
        <li>
            <h3>Update Bridges</h3>
            <p>
                After removing an edge in any iteration, the graph is updated.
                <br>
                Recalculate the bridges in the updated graph to ensure that you have an accurate list of bridges.
            </p>
        </li>
        <li>
            <h3>Repeat</h3>
            <p>
                Repeat the iterative process:
                <br>
                - Find the edge with the maximum weight in the updated graph.
                <br>
                - Check if it is a bridge.
                <br>
                - Remove non-bridge edges until no more can be removed.
                <br>
                - Update the list of bridges.
                <br>
                - Continue this loop until no non-bridge edges can be removed in an iteration.
            </p>
        </li>
        <li>
            <h3>Final Result</h3>
            <p>
                The resulting graph consists of bridge edges, which form the Minimum Spanning Tree of the original graph while ensuring that no non-bridge edges are included.
                <br>
                This MST represents the optimal connected subgraph of the original graph with the maximum-weight edges among non-bridge edges removed.
            </p>
        </li>
    </ol>

<h3> TIME COMPLEXITY </h3>
First lets find the time it takes to identify bridge<br/>
- The time complexity of DFS is O(V + E), where V is the number of vertices, and E is the number of edges.<br/>
- In the worst case, this algorithm visits all vertices and edges once.<br/>
- Therefore, the time complexity of the bridge detection part is O(V + E).<br/>
Now we construct MST by removing non bridge vertices<br/>
- The MST construction part uses a while loop that runs until there are no non-bridge edges left.<br/>
- In each iteration of the loop, the code iterates through all vertices (V) and their adjacent edges (E), which takes O(V + E) time.<br/>
- Since the while loop can run a maximum of V-1 times (once for each vertex in the MST), the overall time complexity for MST construction is O((V + E) * V) = O(V^2 + VE).<br/>

So on comparing we get final complexity of O(V^2 + VE)

````python
from collections import defaultdict

class Graph:

    def __init__(self, vertices):
        self.V = vertices  # Number of vertices
        self.graph = defaultdict(list)  # Default dictionary to store graph
        self.Time = 0

    # Function to add an edge to the graph with a weight
    def addEdge(self, u, v, w):
        self.graph[u].append((v, w))
        self.graph[v].append((u, w))

    # Function to find bridges and store them in a list
    def bridgefind(self, u, visited, parent, low, disc, bridges):
        visited[u] = True
        disc[u] = self.Time
        low[u] = self.Time
        self.Time += 1

        for v, w in self.graph[u]:
            if visited[v] == False:
                parent[v] = u
                self.bridgefind(v, visited, parent, low, disc, bridges)
                low[u] = min(low[u], low[v])

                if low[v] > disc[u]:
                    bridges.append((u, v, w))
            elif v != parent[u]:
                low[u] = min(low[u], disc[v])

    # Function to find bridges and return them in a list
    def bridge(self):
        visited = [False] * (self.V)
        disc = [float("Inf")] * (self.V)
        low = [float("Inf")] * (self.V)
        parent = [-1] * (self.V)
        bridges = []  # List to store bridges

        for i in range(self.V):
            if visited[i] == False:
                self.bridgefind(i, visited, parent, low, disc, bridges)

        return bridges

    # Function to find the MST while excluding non-bridge edges and considering edge weights
    def MST(self):
        mst_edges = []
        while True:
            bridges = self.bridge()
            bridges_set = set((u, v) for u, v, _ in bridges)

            non_bridge_edges = [(u, v, w) for u in range(self.V) for v, w in self.graph[u] if (u, v) not in bridges_set and (v, u) not in bridges_set]

            if not non_bridge_edges:
                break

            max_weight_edge = max(non_bridge_edges, key=lambda x: x[2])
            u, v, w = max_weight_edge

            self.graph[u].remove((v, w))
            self.graph[v].remove((u, w))

        return self.graph

# Driver Code
g = Graph(5)
g.addEdge(0, 1, 10);
g.addEdge(1, 3, 15);
g.addEdge(2, 3, 4);
g.addEdge(2, 0, 6);
g.addEdge(0, 3, 5);

mst_graph = g.MST()

visited = [False] * g.V

def DFS(v):
    visited[v] = True
    for u, w in mst_graph[v]:
        if not visited[u]:
            print(f"{v} - {u} : {w}")
            DFS(u)

for i in range(g.V):
    if not visited[i]:
        DFS(i)
````

Strongly Connected Components (SCCs) are subgraphs within a directed graph where every vertex is reachable from every other vertex within that subgraph. In other words, SCCs represent strongly connected portions of a directed graph.

### Algorithm Used:
To find SCCs in the graph, we follow these steps:

1. Perform Depth-First Search (DFS) on the original graph `G` while keeping track of visited vertices and adding them to a stack in the order they are finished.

2. Transpose the graph `G` to obtain the transpose graph `GT`, where all edges are reversed.

3. Perform DFS on `GT` while popping vertices from the stack generated in step 1. This process identifies SCCs as groups of connected vertices in the reversed graph.

4. Repeat step 3 until the stack is empty.

Using the provided algorithm, we have found the following SCCs in the given graph `G`:

### Conclusion:
In summary, we have successfully identified the Strongly Connected Components in the provided directed graph `G`. The algorithm helps partition the graph into distinct components where each vertex is reachable from every other vertex within the same component. This information can be valuable for various applications such as network analysis and finding connected substructures within complex systems.

````python
def DFS(G, vertex, visited, stack):

    visited[vertex] = True


    for neighbor in G[vertex]:
        if not visited[neighbor]:
            DFS(G, neighbor, visited, stack)


    stack.append(vertex)


def transpose_graph(G):
    GT = {}
    for vertex in G:
        GT[vertex] = []

    for vertex in G:
        for neighbor in G[vertex]:
            GT[neighbor].append(vertex)

    return GT


def find_SCCs(G):
    stack = []
    visited = {}
    for vertex in G:
        visited[vertex] = False

    for vertex in G:
        if not visited[vertex]:
            DFS(G, vertex, visited, stack)

    GT = transpose_graph(G)

    visited = {}
    for vertex in GT:
        visited[vertex] = False

    SCCs = []
    while stack:
        vertex = stack.pop()
        if not visited[vertex]:
            SCC = []
            DFS(GT, vertex, visited, SCC)
            SCCs.append(SCC)

    return SCCs

G = {
    1: [2,4],
    2: [3],
    3: [],
    4: [5],
    5: [6],
    6: [4],
    7: [8,10],
    8: [9],
    9: [10],
    10: [7]
}

SCCs = find_SCCs(G)
print("Strongly Connected Components:")
for SCC in SCCs:
    print(SCC)
````

## Problem Statement
The goal of this code is to determine the maximum threshold value (air quality index) such that if the flights incident to or from the cities with air quality index values less than this threshold are canceled, there should exist a reachable component of cities in the network of size at least K. - `flights`: A list of M tuples where each tuple represents a directed flight from one city to another.

### Output
- `threshold`: The maximum threshold (air quality index) value that satisfies the conditions.

## Code Overview

### 1. Depth-First Search (DFS) Algorithm
The code begins with two DFS functions (`dfs` and `SCC`) that are used to perform a depth-first search on the graph to find strongly connected components. The `dfs` function explores the graph, marking visited nodes and appending them to a stack. The `SCC` function utilizes DFS to find strongly connected components in the graph.

### 2. Finding Strongly Connected Components
The `maxThreshold` function takes the input parameters and initializes the `adj` list to represent the flight network. It then iterates through a binary search loop to find the maximum threshold value.

Inside the loop:
- It checks if there exists a reachable component with air quality greater than or equal to the current `mid` value by using the `SCC` function.
- If such a component is found, it updates `left` to `mid` to search for higher threshold values.
- If not, it updates `right` to `mid - 1` to search for lower threshold values.

The loop continues until `left` is less than or equal to `right`.

### 3. Output
The code returns the maximum threshold value that satisfies the condition described in the problem statement.

The code is tested with the following input:

## Conclusion
The provided code successfully calculates the maximum threshold value based on air quality indices and flights in a city network. It utilizes depth-first search (DFS) to find strongly connected components and binary search to efficiently determine the threshold value that satisfies the specified conditions. The code handles the given problem statement and produces the correct output.

````python
def dfs(v, visited, adj, stack):
    visited[v] = True
    for u in adj[v]:
        if not visited[u]:
            dfs(u, visited, adj, stack)
    stack.append(v)

def SCC(N, adj):
    visited = [False] * N
    stack = []

    for i in range(N):
        if not visited[i]:
            dfs(i, visited, adj, stack)

    graphTranspose = [[] for _ in range(N)]
    for i in range(N):
        for j in adj[i]:
            graphTranspose[j].append(i)

    visited = [False] * N
    scc = []

    while stack:
        v = stack.pop()
        if not visited[v]:
            component = []
            dfs(v, visited, graphTranspose, component)
            scc.append(component)

    return scc

def maxThreshold(N, M, K, aqi, flights):
    adj = [[] for _ in range(N)]
    for u, v in flights:
        adj[u - 1].append(v - 1)

    left = min(aqi)
    right = max(aqi)

    while left < right:
        mid = (left + right + 1) // 2
        canceled = [aqi < mid for aqi in aqi]

        components = SCC(N, adj)

        if any(all(not canceled[node] for node in component) and len(component) >= K for component in components):
            left = mid
        else:
            right = mid - 1

    return (left + right // 2) - 1

N = 6
M = 7
K = 3
aqi = [100, 150, 68, 138, 32, 22]
flights = [[1, 2], [2, 3], [3, 4], [4, 5], [5, 1], [4, 6], [6, 3]]

threshold = maxThreshold(N, M, K, aqi, flights)
print(threshold)
````
