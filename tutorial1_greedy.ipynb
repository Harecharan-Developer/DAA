```python
def fractional_knapsack(items, capacity):
    items.sort(key=lambda x: x[1] / x[0], reverse=True)
    selected_items = []
    total_value = 0

    for item in items:
        if capacity >= item[0]:
            selected_items.append(item)
            capacity -= item[0]
            total_value += item[1]
        else:
            fraction = capacity / item[0]
            selected_items.append((item[0], item[1] * fraction))
            total_value += item[1] * fraction
            break

    return selected_items, total_value

# Example usage:
items = [(2, 10), (3, 5), (5, 15), (7, 7), (1, 6)]
capacity = 10
selected_items, total_value = fractional_knapsack(items, capacity)
print("Selected items:", selected_items)
print("Total value:", total_value)
```

```python
def knapsack_01(items, capacity):
    items.sort(key=lambda x: x[1] / x[0], reverse=True)
    selected_items = []
    current_weight = 0
    total_value = 0

    for item in items:
        if current_weight + item[0] <= capacity:
            selected_items.append(item)
            current_weight += item[0]
            total_value += item[1]

    return selected_items, total_value

# Example usage:
items = [(2, 10), (3, 5), (5, 15), (7, 7), (1, 6)]
capacity = 10
selected_items, total_value = knapsack_01(items, capacity)
print("Selected items:", selected_items)
print("Total value:", total_value)
```

```python
def dijkstra_adjacency_matrix(graph, start):
    num_vertices = len(graph)
    shortest_path = [float('inf')] * num_vertices
    shortest_path[start] = 0
    visited = [False] * num_vertices

    for _ in range(num_vertices):
        min_distance = float('inf')
        min_index = -1

        for v in range(num_vertices):
            if not visited[v] and shortest_path[v] < min_distance:
                min_distance = shortest_path[v]
                min_index = v

        visited[min_index] = True

        for v in range(num_vertices):
            if (
                not visited[v]
                and graph[min_index][v] != 0
                and shortest_path[min_index] != float('inf')
                and shortest_path[min_index] + graph[min_index][v] < shortest_path[v]
            ):
                shortest_path[v] = shortest_path[min_index] + graph[min_index][v]

    return shortest_path

graph_matrix = [
    [0, 2, 0, 6, 0],
    [2, 0, 3, 0, 0],
    [0, 3, 0, 1, 0],
    [6, 0, 1, 0, 3],
    [0, 0, 0, 3, 0],
]

shortest_paths_matrix = dijkstra_adjacency_matrix(graph_matrix, 0)
print("Vertex   Shortest Distance")
for i, distance in enumerate(shortest_paths_matrix):
    print(f"{i}        {distance}")
```

```python
from collections import defaultdict

def dijkstra_adjacency_list(graph, start):
    shortest_path = {}
    for vertex in graph:
        shortest_path[vertex] = float('inf')
    shortest_path[start] = 0
    visited = set()

    while len(visited) < len(graph):
        current_vertex = None
        current_min_distance = float('inf')
        for vertex in shortest_path:
            if shortest_path[vertex] < current_min_distance and vertex not in visited:
                current_min_distance = shortest_path[vertex]
                current_vertex = vertex

        visited.add(current_vertex)

        for neighbor, weight in graph[current_vertex]:
            distance = current_min_distance + weight
            if distance < shortest_path[neighbor]:
                shortest_path[neighbor] = distance

    return shortest_path

# Example usage with adjacency list:
graph_list = {
    'A': [('B', 2), ('D', 6)],
    'B': [('A', 2), ('C', 3)],
    'C': [('B', 3), ('D', 1)],
    'D': [('A', 6), ('C', 1)],
}

shortest_paths_list = dijkstra_adjacency_list(graph_list, 'A')
print("Vertex   Shortest Distance")
for vertex, distance in shortest_paths_list.items():
    print(f"{vertex}        {distance}")
```

```python
def kruskal_adjacency_matrix(graph):
    num_vertices = len(graph)
    result = []
    edges = []

    for i in range(num_vertices):
        for j in range(i + 1, num_vertices):
            if graph[i][j] != 0:
                edges.append((i, j, graph[i][j]))

    edges = sorted(edges, key=lambda edge: edge[2])
    parent = [-1] * num_vertices

    def find(subset, i):
        if subset[i] == -1:
            return i
        return find(subset, subset[i])

    def union(subset, x, y):
        xroot = find(subset, x)
        yroot = find(subset, y)
        subset[xroot] = yroot

    for edge in edges:
        u, v, w = edge
        x = find(parent, u)
        y = find(parent, v)
        if x != y:
            result.append((u, v, w))
            union(parent, x, y)

    return result

# Example usage with adjacency matrix:
graph_matrix = [
    [0, 2, 0, 6, 0],
    [2, 0, 3, 0, 0],
    [0, 3, 0, 1, 0],
    [6, 0, 1, 0, 3],
    [0, 0, 0, 3, 0],
]

minimum_spanning_tree_matrix = kruskal_adjacency_matrix(graph_matrix)
print("Edge   Weight")
for edge in minimum_spanning_tree_matrix:
    u, v, w = edge
    print(f"{u} - {v}     {w}")
```

```python
class Graph:
    def __init__(self):
        self.graph = []

    def add_edge(self, u, v, w):
        self.graph.append((u, v, w))

    def kruskal(self):
        self.graph = sorted(self.graph, key=lambda edge: edge[2])
        parent = {}
        minimum_spanning_tree = []

        def find(node):
            if parent[node] != node:
                parent[node] = find(parent[node])
            return parent[node]

        for u, v, w in self.graph:
            if u not in parent:
                parent[u] = u
            if v not in parent:
                parent[v] = v

            root_u = find(u)
            root_v = find(v)

            if root_u != root_v:
                minimum_spanning_tree.append((u, v, w))
                parent[root_u] = root_v

        return minimum_spanning_tree

# Example usage with adjacency list:
g = Graph()
g.add_edge('A', 'B', 2)
g.add_edge('A', 'D', 6)
g.add_edge('B', 'C', 3)
g.add_edge('C', 'D', 1)

minimum_spanning_tree_list = g.kruskal()
print("Edge   Weight")
for edge in minimum_spanning_tree_list:
    u, v, w = edge
    print(f"{u} - {v}     {w}")
```

prim matrix

```python
def prim_adjacency_matrix(graph):
    num_vertices = len(graph)
    key = [float('inf')] * num_vertices
    parent = [None] * num_vertices
    key[0] = 0
    mst_set = [False] * num_vertices

    for _ in range(num_vertices):
        min_key = float('inf')
        min_index = -1

        for v in range(num_vertices):
            if not mst_set[v] and key[v] < min_key:
                min_key = key[v]
                min_index = v

        mst_set[min_index] = True

        for v in range(num_vertices):
            if (
                graph[min_index][v] != 0
                and not mst_set[v]
                and graph[min_index][v] < key[v]
            ):
                key[v] = graph[min_index][v]
                parent[v] = min_index

    return parent

# Example usage with adjacency matrix:
graph_matrix = [
    [0, 2, 0, 6, 0],
    [2, 0, 3, 0, 0],
    [0, 3, 0, 1, 0],
    [6, 0, 1, 0, 3],
    [0, 0, 0, 3, 0],
]

parent_matrix = prim_adjacency_matrix(graph_matrix)
print("Edge   Weight")
for i in range(1, len(graph_matrix)):
    print(f"{parent_matrix[i]} - {i}     {graph_matrix[i][parent_matrix[i]]}")
```

prim list

```python
class Graph:
    def __init__(self):
        self.graph = {}

    def add_edge(self, u, v, w):
        if u not in self.graph:
            self.graph[u] = []
        if v not in self.graph:
            self.graph[v] = []
        self.graph[u].append((v, w))
        self.graph[v].append((u, w))

    def prim(self):
        num_vertices = len(self.graph)
        key = {vertex: float('inf') for vertex in self.graph}
        parent = {vertex: None for vertex in self.graph}
        key[list(self.graph.keys())[0]] = 0
        mst_set = set()

        for _ in range(num_vertices):
            min_key = float('inf')
            min_vertex = None

            for vertex in self.graph:
                if vertex not in mst_set and key[vertex] < min_key:
                    min_key = key[vertex]
                    min_vertex = vertex

            mst_set.add(min_vertex)

            for neighbor, weight in self.graph[min_vertex]:
                if neighbor not in mst_set and weight < key[neighbor]:
                    key[neighbor] = weight
                    parent[neighbor] = min_vertex

        return parent

# Example usage with adjacency list:
g = Graph()
g.add_edge('A', 'B', 2)
g.add_edge('A', 'D', 6)
g.add_edge('B', 'C', 3)
g.add_edge('C', 'D', 1)

parent_list = g.prim()
print("Edge   Weight")
for vertex, parent in parent_list.items():
    if parent is not None:
        weight = [edge[1] for edge in g.graph[vertex] if edge[0] == parent][0]
        print(f"{parent} - {vertex}     {weight}")
```
