```python
import networkx as nx
import matplotlib.pyplot as plt

class Graph:
    def __init__(self, vertices):
        self.V = vertices
        self.graph = []

    def add_edge(self, u, v, w):
        self.graph.append([u, v, w])

    def find_parent(self, parent, i):
        if parent[i] == i:
            return i
        return self.find_parent(parent, parent[i])

    def union(self, parent, rank, x, y):
        x_root = self.find_parent(parent, x)
        y_root = self.find_parent(parent, y)

        if rank[x_root] < rank[y_root]:
            parent[x_root] = y_root
        elif rank[x_root] > rank[y_root]:
            parent[y_root] = x_root
        else:
            parent[y_root] = x_root
            rank[x_root] += 1

    def edge_marking_mst(self):
        result = []

        # Sort the graph edges by weight in descending order.
        self.graph = sorted(self.graph, key=lambda item: item[2], reverse=True)

        parent = [i for i in range(self.V)]
        rank = [0] * self.V

        while len(result) < self.V - 1:
            u, v, w = self.graph.pop()

            x = self.find_parent(parent, u)
            y = self.find_parent(parent, v)

            if x != y:
                result.append([u, v, w])
                self.union(parent, rank, x, y)

        return result

    def visualize_graph(self):
        G = nx.Graph()
        for u, v, w in self.graph:
            G.add_edge(u, v, weight=w)

        pos = nx.spring_layout(G)  # Layout for the visualization
        labels = nx.get_edge_attributes(G, 'weight')
        nx.draw(G, pos, with_labels=True, node_size=700, node_color='skyblue')
        nx.draw_networkx_edge_labels(G, pos, edge_labels=labels)
        plt.title("Original Graph")
        plt.show()

        mst_edges = self.edge_marking_mst()
        mst_graph = nx.Graph()
        for u, v, w in mst_edges:
            mst_graph.add_edge(u, v, weight=w)

        pos = nx.spring_layout(mst_graph)  # Layout for the visualization
        labels = nx.get_edge_attributes(mst_graph, 'weight')
        nx.draw(mst_graph, pos, with_labels=True, node_size=700, node_color='lightgreen')
        nx.draw_networkx_edge_labels(mst_graph, pos, edge_labels=labels)
        plt.title("Minimum Spanning Tree (MST)")
        plt.show()

# Example usage:
if __name__ == '__main__':
    g = Graph(4)
    g.add_edge(0, 1, 10)
    g.add_edge(0, 2, 6)
    g.add_edge(0, 3, 5)
    g.add_edge(1, 3, 15)
    g.add_edge(2, 3, 4)

    g.visualize_graph()
```

```python
import networkx as nx
import matplotlib.pyplot as plt

class Graph:
    def __init__(self):
        self.graph = {}

    def add_edge(self, u, v):
        if u not in self.graph:
            self.graph[u] = []
        self.graph[u].append(v)

    def get_adjacent_nodes(self, u):
        return self.graph.get(u, [])

    def get_all_nodes(self):
        return self.graph.keys()


class Kosaraju:
    def __init__(self):
        self.visited = set()
        self.stack = []

    def dfs(self, graph, u):
        self.visited.add(u)
        for v in graph.get_adjacent_nodes(u):
            if v not in self.visited:
                self.dfs(graph, v)
        self.stack.append(u)

    def dfs_transpose(self, graph, u, scc):
        self.visited.add(u)
        scc.append(u)
        for v in graph.get_adjacent_nodes(u):
            if v not in self.visited:
                self.dfs_transpose(graph, v, scc)

    def find_scc(self, graph):
        self.visited = set()
        self.stack = []

        for node in graph.get_all_nodes():
            if node not in self.visited:
                self.dfs(graph, node)

        transposed_graph = self.transpose_graph(graph)
        self.visited = set()
        scc_list = []

        while self.stack:
            node = self.stack.pop()
            if node not in self.visited:
                scc = []
                self.dfs_transpose(transposed_graph, node, scc)
                scc_list.append(scc)

        return scc_list

    def transpose_graph(self, graph):
        transposed = Graph()
        for u in graph.get_all_nodes():
            for v in graph.get_adjacent_nodes(u):
                transposed.add_edge(v, u)
        return transposed


# Create a graph
graph = Graph()
graph.add_edge(1, 2)
graph.add_edge(2, 3)
graph.add_edge(1, 4)
graph.add_edge(4, 5)
graph.add_edge(5, 6)
graph.add_edge(6, 4)
graph.add_edge(4, 7)
graph.add_edge(7, 8)
graph.add_edge(8, 9)
graph.add_edge(9, 10)
graph.add_edge(10, 7)

# Find strongly connected components
kosaraju = Kosaraju()
sccs = kosaraju.find_scc(graph)

# Print strongly connected components
print("Strongly Connected Components:")
for i, scc in enumerate(sccs):
    print(f"Component {i + 1}: {scc}")

# Create a NetworkX graph for visualization
G = nx.DiGraph()
for scc in sccs:
    for u in scc:
        for v in graph.get_adjacent_nodes(u):
            if v in scc:
                G.add_edge(u, v)

# Visualize the graph
pos = nx.spring_layout(G)
colors = ['red', 'green', 'blue', 'yellow', 'orange']  # Define colors for each component

nx.draw(G, pos, with_labels=True, node_size=500, node_color='lightblue')
for i, scc in enumerate(sccs):
    nx.draw_networkx_nodes(G, pos, nodelist=scc, node_color=colors[i % len(colors)])
nx.draw_networkx_edges(G, pos, edgelist=G.edges(), edge_color='gray')
nx.draw_networkx_labels(G, pos)
plt.show()
```

```python
def tarjan_scc(N, adj):
    def dfs(v, stack):
        nonlocal index
        index += 1
        low[v] = index
        dfn[v] = index
        stack.append(v)
        on_stack[v] = True

        for u in adj[v]:
            if dfn[u] == 0:
                dfs(u, stack)
                low[v] = min(low[v], low[u])
            elif on_stack[u]:
                low[v] = min(low[v], dfn[u])

        if low[v] == dfn[v]:
            component = []
            while True:
                u = stack.pop()
                on_stack[u] = False
                component.append(u)
                if u == v:
                    break
            sccs.append(component)

    dfn = [0] * N
    low = [0] * N
    on_stack = [False] * N
    stack = []
    sccs = []
    index = 0

    for v in range(N):
        if dfn[v] == 0:
            dfs(v, stack)

    return sccs

def maxThreshold(N, M, K, aqi, flights):
    adj = [[] for _ in range(N)]
    
    for u, v in flights:
        adj[u - 1].append(v - 1)

    left, right = min(aqi), max(aqi)
    result = -1

    while left <= right:
        mid = (left + right) // 2
        canceled = [aqi_val < mid for aqi_val in aqi]

        sccs = tarjan_scc(N, adj)

        if any(len(scc) >= K and all(not canceled[node] for node in scc) for scc in sccs):
            result = mid
            left = mid + 1
        else:
            right = mid - 1

    return result

N = 6
M = 7
K = 3
aqi = [100, 150, 68, 138, 32, 22]
flights = [[1, 2], [2, 3], [3, 4], [4, 5], [5, 1], [4, 6], [6, 3]]

threshold = maxThreshold(N, M, K, aqi, flights)
print(threshold)
```
